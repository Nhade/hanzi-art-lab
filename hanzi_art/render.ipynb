```python
from PIL import Image, ImageDraw, ImageFont
from matplotlib import pyplot as plt
import cv2
```

```python
import json

with open("result.txt", 'r', encoding='UTF-8') as file:
    text = file.readlines()

with open("normalized_color_map.txt", 'r') as file:
    normalized_colors = json.load(file)

fontsize = 16
img = Image.new('RGB', size=(3584*2, 2688*2), color='white')
draw = ImageDraw.Draw(img)
font = ImageFont.truetype('kaiu.ttf', fontsize)

for i in range(len(text)-1):
    for j in range(len(text[i])-1):
        draw.text((j*(fontsize-2), i*(fontsize-2)), text[i][j], font=font, fill=tuple(normalized_colors[i][j]))

# plt.imshow(img)
```

```python
img.save("colorized_render_16px_-2gap_boostx1.3_0.55l.jpg")
```
