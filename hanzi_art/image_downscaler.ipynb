```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread(r'img.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert it to RGB channel
# cropped_img = img[:,:3912]
plt.imshow(img)
```

```python
downscaled_img = cv2.resize(img, dsize=(0,0) ,fx=1/4, fy=1/4)
cv2.imwrite("downscaled.jpg",cv2.cvtColor(downscaled_img, cv2.COLOR_RGB2BGR))
plt.imshow(downscaled_img)
```
